## Контрольные вопросы
**1.	В чем состоит принцип прогнозирования на основе авторегрессии?**
Авторегрессионная модель основана на выражении прогнозируемого значения ряда в момент времени $n > m$ через его известные значения в предыдущие моменты времени: 
<h4 align = "center">$\tilde{x_n} = \sum_{k=1}^{p}ω_kx_{n-p+k-1} $ <h4>
где р — размер «окна» данных, по которому производится прогноз;  — некоторые весовые коэффициенты. Выражение используется для центрированных временных рядов, среднее значение которых равно 0. 
Ошибка прогноза (локальная) равна апостериорной разнице спрогнозированного и реального значений временного ряда:
<h4 align = "center"> $δ_n = x_n - \tilde{x}_n$ <h4>

**2. Объясните методику обучения НС прогноза.**

В режиме обучения последовательно, от эпохи к эпохе, на вход сети подаются элементы векторов-столбцов обучающей выборки:

$$\begin{pmatrix} x_1 & x_2 & ... & x_{m-p} \\ x_2 & x_3 & ... & x_{m-p+1} \\ ... & ... & ... & ... \\ x_p & x_{p+1} & ... & x_{m-1} \end{pmatrix}$$

а на выходе соответственно получаются прогнозируемые значения:
<h4 align = "center"> $\tilde{x}_{p+1}, \tilde{x}_{p+2}, ... , \tilde{x}_{m-1}$ <h4>
Эти значения сравниваются с реальными
<h4 align = "center"> $x_{p+1}, x_{p+2}, ... , x_{m-1}$ <h4>
и по формуле <h4 align = "center"> $δ_n = x_n - \tilde{x}_n$ <h4> оценивается ошибка. Коррекция весов на каждой эпохе производится по правилу Видроу — Хоффа:
<h4 align = "center"> $𝚫ω_k = ŋδx_k, k = \overline{0,p} $ <h4>, где $ŋ$ - норма обучения $ŋ 𝜖(0,1]$
Если по достижении правого края выборки суммарная среднеквадратичная ошибка 
<h4 align = "center"> $ε = \sqrt{\sum_i[x(t_i)-\tilde{x}(t_i)]^2} $ <h4>
останется достаточно большой, следует продолжить обучение, снова вернувшись к первому столбцу выборки и т. д.

**3. Поясните принцип функционирования НС прогноза.**

Нахождение неизвестных весовых коэффициентов w_k осуществляется по известной обучающей выборке значений временного ряда $x_i=x(t_i)  (i=(\overline{1,m})$. Для этого необходимо определить окно длиной $p<m$. Далее, начиная с левого края временного ряда, следует прогнозировать его значения в моменты, идущие непосредственно за окном (справа от него). Сравнивая этот прогноз с реальными значениями, можно оценить ошибку <h4 align = "center"> $δ_n = x_n - \tilde{x}_n$ <h4>, а на ее основе — скорректировать весовые коэффициенты $w_k$, например, с помощью методов обучения НС.

Функция активации — линейная, т. е. $f(net)=net$, где $net$ — комбинированный вход единственного нейрона,
<h4 align = "center"> $net = \sum_{k=1}^{p} ω_kx_{i+k-1}+ω_0$ <h4>
